```python
df = pd.read_parquet('../data/processed/flights_class.parquet')
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

codificador_airline = LabelEncoder()
df['AIRLINE'] = codificador_airline.fit_transform(df['AIRLINE'])    

display(df.head())
```

```python

df['ORIGIN_AIRPORT'] = df['ORIGIN_AIRPORT'].fillna('UNKNOWN').astype(str).str.strip()
codificador_origin_airport = LabelEncoder()
df['ORIGIN_AIRPORT'] = codificador_origin_airport.fit_transform(df['ORIGIN_AIRPORT'])
display(df.head())
```

```python
df['DESTINATION_AIRPORT'] = df['DESTINATION_AIRPORT'].fillna('UNKNOWN').astype(str).str.strip()
codificador_destination_airport = LabelEncoder()
df['DESTINATION_AIRPORT'] = codificador_destination_airport.fit_transform(df['DESTINATION_AIRPORT'])
display(df.head())
```

```python
df2 = df.drop(columns=['ARRIVAL_DELAY'])
y = df2['ARRIVAL_DELAY_CLASS']
x = df2.drop(columns=['ARRIVAL_DELAY_CLASS'])
```

```python
# treinar o modelo de machine learning
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# Criar o modelo de machine learning
modelo_arvore_decisao = RandomForestClassifier(random_state=42, n_jobs = 6)
#modelo_knn = KNeighborsClassifier()

# Treinar o modelo de machine learning
modelo_arvore_decisao.fit(x_train, y_train)
```
